```python
###################
## Initial Setup ##
###################
import pandas as pd
import numpy as np
```

```python
########################
## Load House Dataset ##
########################
df_house = pd.read_excel("../Web_scrapping/df_house.xlsx")

###############################
## clean house dataset       ##
###############################
# # Clean Price Column #
df_house.price = df_house.price.str.replace('$','')
df_house.price = df_house.price.str.replace(',','')
#Clean lot Column
df_house.lot = df_house.lot.str.replace(',','')
df_house.lot = df_house.lot.str.replace('Sq. Ft','')
df_house.lot = df_house.lot.str.replace('—','')
df_house.lot = df_house.lot.str.replace('.','')
df_house.lot = df_house.lot.str.replace('Acres','')
#Clean beds Column
df_house.beds = df_house.beds.str.replace('—','')
#Clean baths Column
df_house.baths = df_house.baths.str.replace('—','')
#Clean Stories Column
df_house.stories = df_house.stories.str.replace('—','')
# Clean lot size column
df_house.lot_size = df_house.lot_size.str.replace(',','')
df_house.lot_size = df_house.lot_size.str.replace('Sq. Ft','')
df_house.lot_size = df_house.lot_size.str.replace('.','')
df_house.lot_size = df_house.lot_size.str.replace('Acres','')
# Clean time on redfin Column
df_house.time_on_redfin = df_house.time_on_redfin.str.replace('days','')
#Clean price per sqft
df_house.price_per_sqft = df_house.price_per_sqft.str.replace('$','')
df_house.price_per_sqft = df_house.price_per_sqft.str.replace('—','')
# Clean view count column
df_house.view_count = df_house.view_count.str.replace(',','')
# Clean renovated year Column
df_house.renovated_year = df_house.renovated_year.str.replace('—','')
# Clean median sale to list 30d Column
df_house.median_list_30d = df_house.median_list_30d.str.replace('$','')
df_house.median_list_30d = df_house.median_list_30d.str.replace('K','000')
# Clean median price per sqft 30d Column
df_house.median_price_per_sqft_30d = df_house.median_price_per_sqft_30d.str.replace('$','')
# Clean median sale price yoy
df_house.median_sale_price_yoy = df_house.median_sale_price_yoy.str.replace('$','')
df_house.median_sale_price_yoy = df_house.median_sale_price_yoy.str.replace(',','')


#added by Brian
df_house.sqft = df_house.sqft.str.replace('—','')
df_house.sqft = df_house.sqft.str.replace(',','')

df_house.address = df_house.address.str.replace(',','')
df_house['city'] = df_house.address_extended.str.extract(r'([A-z ]+)')
df_house['state'] = df_house.address_extended.str.extract(r'(?<=, )([A-z]+)')
df_house['zip'] = df_house.address_extended.str.extract(r'(\d+)')

df_house.median_sale_to_list_30d = df_house.median_sale_to_list_30d.str.replace('%', '')
df_house.median_sale_trend_yoy = df_house.median_sale_trend_yoy.str.replace('+','')
df_house.median_sale_trend_yoy = df_house.median_sale_trend_yoy.str.replace('% year-over-year','')
df_house.homes_sold_trend_yoy = df_house.homes_sold_trend_yoy.str.replace('+','')
df_house.homes_sold_trend_yoy = df_house.homes_sold_trend_yoy.str.replace('% year-over-year','')

df_house.days_on_market_trend_yoy = df_house.days_on_market_trend_yoy.str.extract(r'(\d+)')

df_house['multiple_offers'] = df_house.apply(lambda x: 'Some homes get multiple offers' in str(x.area_facts), axis=1)
df_house['waived_contingencies'] = df_house.apply(lambda x: 'waived contingencies' in str(x.area_facts), axis=1)

df_house['sell_vs_list'] = df_house.area_facts.str.extract(r'(?<=The average homes sell for about )(\d+)')
df_house.sell_vs_list = df_house.sell_vs_list.fillna(0)


df_house['hot_sell_vs_list'] = df_house.area_facts.str.extract(r'(?<=Hot homes can sell for about )(\d+)')
df_house.hot_sell_vs_list = df_house.sell_vs_list.fillna(0)

# Drop property details Column
df_house.drop('property_details',axis=1,inplace=True)
df_house.drop('address_extended',axis=1,inplace=True)
df_house.drop('updated',axis=1,inplace=True)
df_house.drop('area_facts',axis=1,inplace=True)


########################
## Export House Dataset ##
########################
df_house.to_excel('../../data/housing_data_clean.xlsx', index=False)

```

```python

########################
##Load Schools Dataset##
######################## 
schools = pd.read_excel("../Web_scrapping/df_schools.xlsx")

##############################
## Cleaning Schools Dataset ##
##############################
# Clean overall rating Column
schools.overall_rating = schools.overall_rating.str.replace('-','')
# Clean student count Column
schools.student_count = schools.student_count.str.replace('-','')
# Clean school distance column
schools.school_distance = schools.school_distance.str.replace('mi','')
#Clean review count column
schools.review_count = schools.review_count.str.replace('reviews','')
#Clean review rating column
schools.review_rating = schools.review_rating.str.replace('Rating','')
schools.review_rating = schools.review_rating.str.replace('out of 5','')
schools.review_rating = schools.review_rating.str.replace('.0','')

########################
## Export House Dataset ##
########################
schools.to_excel('../../data/schools_data_clean.xlsx', index=False)
```

```python
###############################
## Load Transactions Dataset ##
###############################
transactions = pd.read_excel("../Web_scrapping/df_transactions.xlsx")

###################################
## Cleaning Transactions Dataset ##
###################################
# Clean transaction dates column
transactions['transaction_dates'] = pd.to_datetime(transactions['transaction_dates'],infer_datetime_format=True)
#Clean transaction amount Column
transactions.transaction_amount = transactions.transaction_amount.str.replace(',','')
transactions.transaction_amount = transactions.transaction_amount.str.extract(r'(\d+)')

########################
## Export House Dataset ##
########################
transactions.to_excel('../../data/transactions_data_clean.xlsx', index=False)
```
